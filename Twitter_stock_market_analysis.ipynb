# **Let's analyze stock prices of Twitter of the years**

```python
#Date	Open	High	Low	Close
figure = go.Figure(data=[go.Candlestick(x=data["Date"],open=data["Open"],high=data["High"],low=data["Low"],close=data["Close"])])
figure.update_layout(title="Twitter stock price analysis over all years")
figure.show()
#range slider works here by default
```

# **Assigning buttons to control Time Periods.Let's add buttons to analyze the stock prices of Twitter in different Time Periods:**

```python
figure=px.bar(data,x="Date",y="Close",color="Close")
figure.update_xaxes(rangeslider_visible=True) #the below graph that makes things selected 
figure.update_layout(title="Twitter Stock Prices Over the Years",xaxis_rangeslider_visible=True)

figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1,label="1m",step="month",stepmode="backward"),
            dict(count=3,label="3m",step="month",stepmode="backward"),
            dict(count=6,label="6m",step="month",stepmode="backward"),
            dict(count=1,label="1y",step="year",stepmode="backward"),
            dict(count=2,label="2y",step="year",stepmode="backward"),
            dict(step="all")
        ])
    )
)
figure.show()
```
